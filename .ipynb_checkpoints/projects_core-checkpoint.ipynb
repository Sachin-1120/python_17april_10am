```python
import secrets
import string

pw_length = int(input("Enter the desired password length: "))
letters = string.ascii_letters
digits = string.digits
special_chars = string.punctuation

alphabet = letters + digits + special_chars
pwd = ''
pw_strong = False

while not pw_strong:
    pwd = ''
    for i in range(pw_length):
        pwd += ''.join(secrets.choice(alphabet))

    if (any(char in special_chars for char in pwd) and
       sum(char in digits for char in pwd) >= 2):
        pw_strong = True

if pw_length > 8:
    print("Password is strong:", pwd)
else:
    print("Password is good:", pwd)
```

```python
import random
import os
import re


def check_play_status():
    valid_responses = ['yes', 'no']
    while True:
        try:
            response = input('Do you wish to play again? (Yes or No): ')
            if response.lower() not in valid_responses:
                raise ValueError('Yes or No only')

            if response.lower() == 'yes':
                return True
            else:
                os.system('cls' if os.name == 'nt' else 'clear')
                print('Thanks for playing!')
                exit()

        except ValueError as err:
            print(err)


def play_rps():
    play = True
    while play:
        os.system('cls' if os.name == 'nt' else 'clear')
        print('')
        print('Rock, Paper, Scissors - Shoot!')

        user_choice = input('Choose your weapon'
                           ' [R]ock], [P]aper, or [S]cissors: ')

        if not re.match("[SsRrPp]", user_choice):
            print('Please choose a letter:')
            print('[R]ock, [P]aper, or [S]cissors')
            continue

        print(f'You chose: {user_choice}')

        choices = ['R', 'P', 'S']
        opp_choice = random.choice(choices)

        print(f'I chose: {opp_choice}')

        if opp_choice == user_choice.upper():
            print('Tie!')
            play = check_play_status()
        elif opp_choice == 'R' and user_choice.upper() == 'S':
            print('Rock beats scissors, I win!')
            play = check_play_status()
        elif opp_choice == 'S' and user_choice.upper() == 'P':
            print('Scissors beats paper! I win!')
            play = check_play_status()
        elif opp_choice == 'P' and user_choice.upper() == 'R':
            print('Paper beats rock, I win!')
            play = check_play_status()
        else:
            print('You win!\n')
            play = check_play_status()


if __name__ == '__main__':
    play_rps()
```

# File handling

```python
import os
import hashlib

def get_file_checksum(file_path, block_size=65536):
    """Calculate the checksum of a file using SHA-256."""
    hasher = hashlib.sha256()
    with open(file_path, 'rb') as file:
        for block in iter(lambda: file.read(block_size), b''):
            hasher.update(block)
    return hasher.hexdigest()

def find_duplicate_files(directory):
    """Find and return a list of duplicate files in the specified directory."""
    hash_dict = {}
    duplicates = []

    for root, dirs, files in os.walk(directory):
        for filename in files:
            file_path = os.path.join(root, filename)
            file_hash = get_file_checksum(file_path)

            if file_hash in hash_dict:
                duplicates.append((file_path, hash_dict[file_hash]))
            else:
                hash_dict[file_hash] = file_path

    return duplicates

def remove_duplicate_files(duplicate_list):
    """Remove duplicate files."""
    for duplicate_pair in duplicate_list:
        print(f"Removing duplicate: {duplicate_pair[0]}")
        os.remove(duplicate_pair[0])

# Example usage
directory_path = input("Enter the directory path to search for duplicate files: ")

duplicate_files = find_duplicate_files(directory_path)

if duplicate_files:
    print("Duplicate files found:")
    for duplicate_pair in duplicate_files:
        print(f"{duplicate_pair[0]} duplicates with {duplicate_pair[1]}")

    remove_duplicates = input("Do you want to remove duplicate files? (yes/no): ").lower()
    if remove_duplicates == 'yes':
        remove_duplicate_files(duplicate_files)
        print("Duplicate files removed.")
    else:
        print("No files were removed.")
else:
    print("No duplicate files found.")
```

```python
import datetime
import time
import pygame

def set_alarm():
    print("Set the alarm time (24-hour format HH:MM):")
    while True:
        try:
            alarm_time_str = input("Enter the time: ")
            alarm_time = datetime.datetime.strptime(alarm_time_str, "%H:%M")
            break
        except ValueError:
            print("Invalid time format. Please use HH:MM.")

    return alarm_time

def play_alarm_sound():
    pygame.mixer.init()
    pygame.mixer.music.load("clock-alarm-8761.mp3")  # Provide the path to your alarm sound file
    pygame.mixer.music.play()
    time.sleep(10)  # Adjust the sleep duration based on your alarm sound length
    pygame.mixer.music.stop()

def main():
    print("Welcome to the Alarm Clock!")

    alarm_time = set_alarm()

    while True:
        current_time = datetime.datetime.now().strftime("%H:%M")
        print(f"Current time: {current_time}")

        if current_time == alarm_time.strftime("%H:%M"):
            print("Time to wake up!")
            play_alarm_sound()
            break

        time.sleep(60)  # Check the time every minute

if __name__ == "__main__":
    main()
```

```python
import datetime

# Dictionary to store birthdays (replace with your own data)
birthdays = {
    'Alice': '1990-01-19',
    'Bob': '1985-08-22',
    'Charlie': '1995-02-10',
    # Add more entries as needed
}

while True:
    print("\nBirthday Reminder")
    print("1. Check upcoming birthdays")
    print("2. Exit")

    choice = input("Enter your choice (1 or 2): ")

    if choice == '1':
        today = datetime.date.today()
        upcoming_birthdays = []

        for name, dob in birthdays.items():
            birthdate = datetime.datetime.strptime(dob, '%Y-%m-%d').date()

            if birthdate.month == today.month and birthdate.day >= today.day:
                upcoming_birthdays.append((name, birthdate))

        if upcoming_birthdays:
            print("\nUpcoming Birthdays:")
            for name, dob in upcoming_birthdays:
                print(f"{name}: {dob.strftime('%B %d')}")
        else:
            print("\nNo upcoming birthdays in the next month.")

    elif choice == '2':
        print("Exiting the birthday reminder. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter 1 or 2.")
```

```python
import datetime
import time

def add_task(tasks, description, deadline):
    task = {"description": description, "deadline": deadline}
    tasks.append(task)
    print(f"Task '{description}' added with a deadline of {deadline}.\n")

def show_tasks(tasks):
    if not tasks:
        print("No tasks available.")
    else:
        print("Tasks:")
        for idx, task in enumerate(tasks, start=1):
            print(f"{idx}. {task['description']} - Deadline: {task['deadline']}")
        print()

def remove_task(tasks, task_index):
    if 1 <= task_index <= len(tasks):
        removed_task = tasks.pop(task_index - 1)
        print(f"Task '{removed_task['description']}' removed.\n")
    else:
        print("Invalid task index.\n")

def update_task(tasks, task_index, description, deadline):
    if 1 <= task_index <= len(tasks):
        tasks[task_index - 1]['description'] = description
        tasks[task_index - 1]['deadline'] = deadline
        print(f"Task updated successfully.\n")
    else:
        print("Invalid task index.\n")

def set_reminder(task, reminder_time):
    current_time = datetime.datetime.now()
    deadline = datetime.datetime.strptime(task['deadline'], "%Y-%m-%d %H:%M:%S")

    time_difference = (deadline - current_time).total_seconds()
    reminder_delay = (reminder_time - current_time).total_seconds()

    if 0 <= time_difference <= reminder_delay:
        print(f"Reminder: Task '{task['description']}' is due soon!")

def main():
    tasks = []
    
    while True:
        print("1. Add Task")
        print("2. Show Tasks")
        print("3. Remove Task")
        print("4. Update Task")
        print("5. Exit")

        choice = input("Enter your choice (1/2/3/4/5): ")

        if choice == "1":
            description = input("Enter task description: ")
            deadline_str = input("Enter task deadline (YYYY-MM-DD HH:MM:SS): ")

            try:
                deadline = datetime.datetime.strptime(deadline_str, "%Y-%m-%d %H:%M:%S")
            except ValueError:
                print("Invalid deadline format. Please use YYYY-MM-DD HH:MM:SS.")
                continue

            add_task(tasks, description, deadline)

            # Set a reminder 5 minutes before the task deadline
            reminder_time = deadline - datetime.timedelta(minutes=5)
            set_reminder(tasks[-1], reminder_time)

        elif choice == "2":
            show_tasks(tasks)

        elif choice == "3":
            task_index = int(input("Enter the task index to remove: "))
            remove_task(tasks, task_index)

        elif choice == "4":
            task_index = int(input("Enter the task index to update: "))
            description = input("Enter updated task description: ")
            deadline_str = input("Enter updated task deadline (YYYY-MM-DD HH:MM:SS): ")

            try:
                deadline = datetime.datetime.strptime(deadline_str, "%Y-%m-%d %H:%M:%S")
            except ValueError:
                print("Invalid deadline format. Please use YYYY-MM-DD HH:MM:SS.")
                continue

            update_task(tasks, task_index, description, deadline)

        elif choice == "5":
            print("Exiting program.")
            break

        else:
            print("Invalid choice. Please enter 1, 2, 3, 4, or 5.")

if __name__ == "__main__":
    main()
```
