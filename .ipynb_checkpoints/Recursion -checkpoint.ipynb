```python
# factorial 
n=int(input('Enter number'))
fact=1
for i in range(1,n+1):
    fact*=i
print('Fatorial',fact)
```

```python
def fact(n):
    if n==1:
        return 1
    else:
        return n*fact(n-1)
#                5*fact(4)
fact(5)
```

```python
def fact(n):
    if n==3:
        return 6
    return n*fact(n-1)
fact(5)
```

```python
def palindrome(n):
    if len(n)<=1:
           return 'Palindrome'
    if n[0]==n[-1]:
        return palindrome(n[1:-1])
    else:
        return 'Not Palindrome'
palindrome(input('Enter String').lower())
```

```python
def prime(n,i=2):
    if i==n:
        return 'Prime'
    if n%i==0:
        return 'Not Prime'
    else:
        return prime(n,i+1)
prime(5)
```

```python
def print_alpha(n,i=1):
    if i==10:
        return f'{n} * {i} = {n*i}'
    print(f'{n} * {i} = {n*i}')
    return print_alpha(n,i+1)
print(print_alpha(int(input('Enter Number'))))
```

```python
#recursion 
#l=[1,12,3,4,4,66,7]
def sum_alpha(n,i=0,sum_i=0):
    if i==len(n):
        return sum_i
    return sum_alpha(n,i+1,sum_i+n[i])
sum_alpha([1,12,3,4,4,66,7])
```
